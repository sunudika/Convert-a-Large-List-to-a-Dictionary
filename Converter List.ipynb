```python
DATA_KBBI 	= []
DATA_KBBI	= [kamus.strip('\n').strip('\r') for kamus in open('kamus/kbba.txt')]

dic={}
for i in DATA_KBBI: 
    (key,val)=i.split('\t')
    dic[str(key)]=val
```

```python
dic
```

```python
with open('file.txt', 'w') as file:
    for k in sorted (dic.keys()):
        file.write("'%s':%s', \n" % (k, dic[k]))
    file.write("}")
```
